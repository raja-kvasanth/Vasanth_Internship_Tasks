## **Running sum of 1D Array**

```python
def running_sum(nums):
  for i in range(1,len(nums)):
    nums[i]=nums[i]+nums[i-1]
  return nums
op=running_sum([1,1,1,1,1])
print(op)
```

## **Shuffle the array**

```python
def shuffle(nums,n):
  output=[]
  for i in range(n):
    output.append(nums[i])
    output.append(nums[i+n])
  return output
op=shuffle([2,5,1,3,4,7],3)
print(op)
```

## **Kids With the Greatest Number of Candies**

```python
def chk_Greatest_Candies(candies,extraCandies):
  max_Candies=max(candies)
  output=[]
  for i in candies:
    if i+extraCandies>=max_Candies:
      output.append(True)
    else:
      output.append(False)
  return output
check=chk_Greatest_Candies([2,3,5,1,3],3)
print(check)
```
