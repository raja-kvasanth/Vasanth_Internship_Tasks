## **Defanging an IP Address**

```python
def ipaddr(address):
  result=''
  for i in address:
    if i=='.':
      result=result+'[.]'
    else:
      result=result+i
  return result
op=ipaddr('255.100.50.0')
print(op)
```

## **Find Numbers with Even Number of Digits**

```python
def find_EvenNum_Digits(nums):
  count=0
  for i in nums:
    if len(str(i))%2==0:
      count=count+1
  return count
op=find_EvenNum_Digits([555,901,482,1771])
print(op)
```

## **Number of Good Pairs**

```python
def goodPair(nums):
  count=0
  for i in nums:
    for j in nums:
      if nums[i]==nums[j]:
        count=count+1
  return count
op=goodPair([1,2,3,1,1,3])
print(op)
```

## **How Many Numbers Are Smaller Than the Current Number**

```python
def smallerNumbers(nums):
      sorted_nums=sorted(nums)
      result=[]

      for n in nums:
          result.append(sorted_nums.index(n))

      return result
op=smallerNumbers([8,1,2,2,3])
print(op)
```

## **Subtract the Product and Sum of Digits of an Integer**

```python
def subtractPrdctNdSum(n):
       product=1
       total=0

       for digit in str(n):
          d=int(digit)
          product*=d
          total+=d

       return product-total
op=subtractPrdctNdSum(234)
print(op)
```

## **XOR Operation in an Array**

```python
def xorOperation(n,start):
       result=0

       for i in range(n):
          result^=(start+2*i)

       return result
op=xorOperation(5,0)
print(op)
```
